# Libraries

```python
import requests
from bs4 import BeautifulSoup
import re
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.corpus import stopwords
```

# Get HTML

```python
url = 'https://en.wikipedia.org/wiki/Lionel_Messi'

response = requests.get(url)
html = response.text
```

# Extract Text

```python
soup = BeautifulSoup(html, 'html.parser')
text = soup.get_text()
```

# Remove puncutations

W --> for any NON-word character

w --> for any word character (a-z,A-Z,0-9,_) 

```python


text = re.sub(r'\W+', ' ', text)
```

# Normalization

```python
text = text.lower()
```

# Split text 

```python
words = text.split()
```

# Lemmatization or Stemming

Return words to its origin

**Stemming**: is a process that removes the endings of words.  
Good: the stem of the words “jumps”, “jumped”, and “jumping” is “jump”.                                     
Bad: the stem of “universe” and “university” would both be “univers”, which are clearly two different words


**Lemmatization**: on the other hand, reduces words to their base or root form                                         
Good: Lemmatization is more complex and accurate than stemming.                                                           
Bad: lemmatization is computationally more expensive than stemming.

```python
lemmatizer = WordNetLemmatizer()
words = [lemmatizer.lemmatize(word) for word in words]  #lemmatization

#stemmer = PorterStemmer()
#words = [stemmer.stem(word) for word in words]  #stemming
```

# Remove stop words

```python
stop_words = set(stopwords.words('english'))
words = [word for word in words if word not in stop_words]
```

# Get unique words

```python
unique_words = set(words)
print(unique_words)
```
